# **O que são Algoritmos de Ordenação?**



São procedimentos computacionais compostos por uma sequência finita e bem definida de operações cujo objetivo é reorganizar um conjunto de elementos de acordo com um critério de ordenação preestabelecido (por exemplo, crescente ou decrescente).

# **Importância dos Algoritmos de Ordenação**



- **Bancos de Dados**: Os dados em um banco de dados frequentemente precisam ser recuperados em uma ordem específica. Algoritmos de ordenação são usados para organizar dados em tabelas de bancos de dados.

- **Aplicações *Web***: Muitas aplicações *web* exibem listas de informações em ordens específicas, como listas de produtos em uma loja *online*. Algoritmos de ordenação garantem que essas informações sejam apresentadas de forma organizada.

- **Processamento de Dados**: Dados brutos frequentemente precisam ser organizados antes de serem processados. Algoritmos de ordenação são usados para preparar os dados para análise subsequente.

- **Algoritmos de Busca Eficiente**: Algoritmos de ordenação são frequentemente usados como parte de algoritmos de busca eficiente, como a busca binária, que requer que os dados estejam ordenados.

# **Métricas de Avaliação**

Ao escolher um algoritmo de ordenação para uma determinada tarefa, várias métricas de avaliação são consideradas:

- **Complexidade Temporal**: Refere-se ao tempo de execução do algoritmo. Algoritmos mais eficientes têm menor complexidade temporal, o que é crucial para ordenar grandes conjuntos de dados.

- **Complexidade Espacial**: Refere-se à quantidade de memória adicional usada pelo algoritmo. Algoritmos com menor complexidade espacial usam menos memória, o que é importante em sistemas com restrições de memória.

- **Estabilidade**: Algoritmos estáveis preservam a ordem relativa dos elementos com chaves iguais. Isso é importante em cenários onde múltiplos critérios de ordenação são aplicados sequencialmente.

# **Métodos de Ordenação**

Os algoritmos de ordenação podem ser classificados em dois grupos, com base na disponibilidade de memória para realizar a ordenação:









## **Ordenação Interna**

Todos os elementos a serem ordenados cabem na memória principal (RAM) do computador. Isso permite que qualquer registro seja acessado rapidamente, tornando a ordenação mais eficiente. Os métodos de ordenação interna podem ser divididos em dois subgrupos:

- **Métodos Simples**: Esses métodos são adequados para pequenos conjuntos de dados e são relativamente fáceis de implementar e entender. No entanto, eles têm complexidade quadrática, o que significa que podem se tornar lentos para grandes volumes de dados. Exemplos incluem o *Bubble Sort*, o *Insertion Sort*, o *Selection Sort* e o *Comb Sort*.

- **Métodos Eficientes**: Esses métodos são mais complexos em termos de detalhes, mas oferecem melhor desempenho para conjuntos de dados maiores. Eles têm uma complexidade logarítmica, tornando-os mais rápidos para grandes volumes de dados. Alguns exemplos populares são o *Quick Sort*, o *Merge Sort*, o *Heap Sort*, o *Shell Sort* e o *TimSort*.

## **Ordenação Externa**

Os elementos a serem ordenados não cabem na memória principal e precisam ser acessados sequencialmente ou em grandes blocos. Isso é comum quando você precisa ordenar grandes volumes de dados que não podem ser carregados completamente na memória RAM. A ordenação externa é mais complexa e envolve leitura e gravação de dados em dispositivos de armazenamento secundário, como discos rígidos. Algoritmos como o *Merge Sort* e o *Polyphase Sort* são usados para lidar com a ordenação externa.

Esse processo é repetido várias vezes até que a lista inteira esteja ordenada. ###***Swap***

A troca de dois elementos numa lista requer o uso de uma variável para armazenamento temporário, conforme trecho de código abaixo:

````python
numeros = [93, 58, 44]

temp = numeros[0]  # Armazena temporariamente o valor 93
numeros[0] = numeros[2]  # Substitui o primeiro valor pelo último: [44, 58, 44]
numeros[2] = temp  # Substitui o último valor pelo temporário (93): [44, 58, 93]

print(numeros)  # Saída: [44, 58, 93]
````

Porém, em *Python*, é possível realizar atribuições simultâneas, conforme exemplo abaixo:

````python
numeros = [93, 58, 44]

# Troca direta entre o primeiro e o último elemento
numeros[0], numeros[2] = numeros[2], numeros[0]

print(numeros)  # Saída: [44, 58, 93]
````

###**Implementação**

#### **Descrição Formal**

Dado um vetor $A[0 \dots n-1]$:

1. **Inicialização:**
   O algoritmo percorre repetidamente o vetor, comparando elementos adjacentes e trocando-os de posição quando estão fora de ordem.

2. **Iteração Externa:**
   Para cada passada $i$ de $0$ até $n-2$, o maior elemento da parte **não ordenada** é movido para sua posição final correta no final do vetor.

   * A cada iteração, a região ordenada ao final do vetor cresce, reduzindo a parte que precisa ser percorrida.

3. **Iteração Interna:**
   Para cada índice $j$ de $0$ até $n-i-2$:

   * Se $A[j] > A[j+1]$, troca-se $A[j]$ com $A[j+1]$.

4. **Otimização (variante adaptativa):**
   Se durante uma passada nenhuma troca ocorrer, o algoritmo pode ser interrompido antecipadamente, pois o vetor já está ordenado.

5. **Complexidade:**

   * Melhor caso ($O(n)$): vetor já ordenado (com otimização).
   * Pior caso ($O(n^2)$): vetor ordenado inversamente.
   * Espaço auxiliar: $O(1)$ (**in-place**).
   * Estável: mantém a ordem relativa de elementos iguais.




#### **Pseudocódigo**

```text
BubbleSort(A, n):
    para i ← 0 até n-2:
        trocou ← falso
        
        para j ← 0 até n-i-2:
            se A[j] > A[j+1]:
                trocar A[j] ↔ A[j+1]
                trocou ← verdadeiro
        
        se trocou = falso:
            interromper
```

#### **Primeira Implementação (não otimizada)**

O *Bubble Sort* é um dos algoritmos de ordenação mais simples e diretos. Ele recebe esse nome porque os elementos "borbulham" através do vetor conforme são comparados e trocados de posição, até que todos os elementos estejam em sua posição correta. Observe abaixo:

````python
def bubble_sort_nao_otimizado(lista):
    """Ordena a lista utilizando Bubble Sort não otimizado."""
    n = len(lista)  # Tamanho da lista
    # Percorre a lista
    for i in range(n):  # Executa n vezes
        for j in range(0, n - 1):  # Executa n - 1 vezes
            # Compara elementos adjacentes
            if lista[j] > lista[j + 1]:
                # Troca os elementos de posição
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
````

#### **Segunda Implementação (otimizada)**

É possível perceber que, a cada iteração, um número é "empurrado" para o final da lista, encontrando a sua posição ordenada. Consequentemente, durante as iterações subsequentes, o índice associado a esse número não necessita mais ser comparado, uma vez que sua posição já está estabelecida. Em termos mais específicos, à medida que o algoritmo progride, o tamanho efetivo da lista a ser considerada diminui, garantindo que cada iteração resulte em um índice a menos que requer análise. Isso é particularmente evidente na otimização que foi implementada no segundo *loop* *for*:

````python
def bubble_sort_otimizado(lista):
    """Ordena a lista utilizando Bubble Sort otimizado."""
    n = len(lista)  # Tamanho da lista
    # Percorre a lista
    for i in range(n):
        # Últimos i elementos já estão ordenados, não precisam ser comparados
        for j in range(0, n - 1 - i):
            # Compara elementos adjacentes
            if lista[j] > lista[j + 1]:
                # Troca os elementos de posição
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
````

#### **Terceira Implementação (recursiva)**

Não é a abordagem mais eficiente para esse algoritmo, uma vez que ele já é ineficiente em sua forma iterativa. No entanto, se você estiver interessado em ver como poderia ser feita uma implementação recursiva do *Bubble Sort*, aqui está um exemplo:



````python
def bubble_sort_recursivo(lista, n):
    """Ordena a lista utilizando Bubble Sort de forma recursiva."""
    # Caso base: lista com 0 ou 1 elemento já está ordenada
    if n <= 1:
        return

    # Percorre a lista comparando elementos adjacentes
    for i in range(n - 1):
        if lista[i] > lista[i + 1]:
            # Troca os elementos de posição
            lista[i], lista[i + 1] = lista[i + 1], lista[i]

    # Chama recursivamente com tamanho reduzido
    bubble_sort_recursivo(lista, n - 1)
````

**Obs**: pode levar ao estouro de pilha (listas maiores), já que uma chamada recursiva é feita para cada passagem pelo conjunto de elementos.

### **Exemplo de Uso**

````python
numeros = [64, 34, 25, 12, 22, 11, 90]
print("Lista original:", numeros)
bubble_sort_otimizado(numeros)
print("Lista ordenada:", numeros)
````

### **Análise do Tempo**

Abaixo está o gráfico comparativo do tempo gasto pelas três diferentes implementações do *Bubble Sort*.

````python
import time
import matplotlib.pyplot as plt

# Gerar listas de tamanhos diferentes para a análise
tamanhos_lista = [100, 200, 300, 400, 500, 600, 700, 800, 900]
tempos_nao_otimizado = []
tempos_otimizado = []
tempos_recursivo = []

# Número de repetições
n_execucoes = 100

for tamanho in tamanhos_lista:
    dados = list(range(tamanho, 0, -1))  # Lista decrescente

    # Bubble Sort Não Otimizado
    tempo_total = 0
    for _ in range(n_execucoes):
        inicio = time.time()
        bubble_sort_nao_otimizado(dados.copy())
        tempo_total += time.time() - inicio
    tempos_nao_otimizado.append(tempo_total / n_execucoes)

    # Bubble Sort Otimizado
    tempo_total = 0
    for _ in range(n_execucoes):
        inicio = time.time()
        bubble_sort_otimizado(dados.copy())
        tempo_total += time.time() - inicio
    tempos_otimizado.append(tempo_total / n_execucoes)

    # Bubble Sort Recursivo
    tempo_total = 0
    for _ in range(n_execucoes):
        inicio = time.time()
        bubble_sort_recursivo(dados.copy(), tamanho)
        tempo_total += time.time() - inicio
    tempos_recursivo.append(tempo_total / n_execucoes)

# Plotar os resultados
plt.plot(tamanhos_lista, tempos_nao_otimizado, marker='o', label='Não Otimizado')
plt.plot(tamanhos_lista, tempos_otimizado, marker='o', label='Otimizado')
plt.plot(tamanhos_lista, tempos_recursivo, marker='o', label='Recursivo')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo Médio (segundos)')
plt.title(f'Comparação de Tempo Médio - Bubble Sort ({n_execucoes} execuções)')
plt.legend()
plt.grid(True)
plt.show()
````

### **Complexidade**

A tabela a seguir apresenta o número de comparações realizadas em cada passagem do algoritmo *Bubble Sort*, considerando um total de $𝑛$ elementos a serem ordenados. Conforme o processo avança, o número de comparações diminui progressivamente, já que a maior parte dos elementos já estará corretamente posicionada. O número total de passagens é $𝑛−1$, e o número de comparações em cada passagem decresce de acordo com os valores descritos. Isso reflete a natureza do *Bubble Sort*, onde a maior parte das trocas necessárias ocorre nas primeiras passagens, diminuindo gradualmente à medida que a lista se torna mais ordenada.

| Passagem | Comparações |
| --- | --- |
| $1$ | $n-1$ |
| $2$ | $n-2$ |
| $3$ | $n-3$ |
| ... | ... |
| $n-1$ | $1$ |

- $𝑛−1$ passagens serão realizadas para ordenar uma lista de tamanho $𝑛$;
- A soma das comparações é uma PA: $𝑛-1 + 𝑛-2 + 𝑛-3 + ... + 3 + 2 + 1$;
- Lembre-se de que essa soma pode ser calculada como: $𝑛²/2 - 𝑛/2$;
  - para $𝑛 = 4$, temos $𝑛²/2 = 8$ e $𝑛/2 = 2$.
  - para $𝑛 = 1000$, temos $𝑛²/2 = 500.000$ e $𝑛/2 = 500$.
- Ou seja, $𝑂(𝑛²)$ comparações.

**Obs**: Embora o *Bubble Sort* seja ineficiente em cenários de grande escala, ele ainda é útil quando o conjunto de dados está quase ordenado. Isso ocorre porque o número de trocas será menor e, em muitos casos, será mais rápido do que outros algoritmos mais sofisticados.

### **Vantagens e Limitações**



#### **Vantagens**:



- Fácil de entender e implementar.
- Útil para ilustrar os conceitos básicos de ordenação.
- Pode ser útil para listas pequenas ou quase ordenadas.

#### **Limitações**:

- Complexidade quadrática no pior caso (não eficiente para grandes conjuntos de dados).
- Muitas comparações e trocas, mesmo em casos ideais.
